```python
import pandas as pd

file_8 = pd.read_csv('../data/traindataset8x8.csv')
file_16 = pd.read_csv('../data/traindataset16x16.csv')
file_32 = pd.read_csv('../data/traindataset32x32.csv')
```

```python
X_train_64=file_8.drop('64',axis=1)
y_train_64=file_8['64']
```

```python
file_16.head()
```

```python
X_train_256=file_16.drop('256',axis=1)
y_train_256=file_16['256']
```

```python
X_train_1024=file_32.drop('1024',axis=1)
y_train_1024=file_32['1024']
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn_model_64 = KNeighborsClassifier(n_neighbors=3)
knn_model_64.fit(X_train_64,y_train_64)
joblib.dump(knn_model_64, 'knn_model_64.pkl')


knn_model_256 = KNeighborsClassifier(n_neighbors=3)
knn_model_256.fit(X_train_256,y_train_256)
joblib.dump(knn_model_256, 'knn_model_256.pkl')


knn_model_1024 = KNeighborsClassifier(n_neighbors=3)
knn_model_1024.fit(X_train_1024,y_train_1024)
joblib.dump(knn_model_1024, 'knn_model_1024.pkl')


# knnmodpred=knnmodel.predict(X_test)
# accuracy = accuracy_score(Y_test, knnmodpred)
# testing_acc.append(accuracy)
# print("Accuracy:", accuracy)

# cm_test = confusion_matrix(Y_test,knnmodpred)
# print(cm_test)
# # TP, TN,FP, FN=conf_spec(cm_test)
# # print_spec(TP,TN,FP,FN,22,"For Test Set:")
# y_predtrain=knnmodel.predict(X_train)
# train_accuracy = accuracy_score(Y_train, y_predtrain)
# training_acc.append(train_accuracy)
# print("Train Accuracy:", train_accuracy)
```

```python
from sklearn.tree import DecisionTreeClassifier

dt_model_64 = DecisionTreeClassifier()
dt_model_256 = DecisionTreeClassifier()
dt_model_1024 = DecisionTreeClassifier()

dt_model_64.fit(X_train_64,y_train_64)
joblib.dump(dt_model_64, 'dt_model_64.pkl')


dt_model_256.fit(X_train_256,y_train_256)
joblib.dump(dt_model_256, 'dt_model_256.pkl')


dt_model_1024.fit(X_train_1024,y_train_1024)
joblib.dump(dt_model_1024, 'dt_model_1024.pkl')
```

```python
from sklearn.svm import SVC

svm_model_64 = SVC(kernel='rbf')
svm_model_64.fit(X_train_64, y_train_64)
joblib.dump(svm_model_64, 'svm_model_64.pkl')


svm_model_256 = SVC(kernel='rbf')
svm_model_256.fit(X_train_256, y_train_256)
joblib.dump(svm_model_256, 'svm_model_256.pkl')


svm_model_1024 = SVC(kernel='rbf')
svm_model_1024.fit(X_train_1024, y_train_1024)
joblib.dump(svm_model_1024, 'svm_model_1024.pkl')

```
